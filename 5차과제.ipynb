```python
#3번 정답
days_of_war=pd.read_csv("revolutionary_war.csv",usecols=["Start Date"],parse_dates=["Start Date"]).squeeze(1)

days_of_war
```

```python
from datetime import datetime

def days_of_week(frame):
    return frame.dt.strftime("%A")

#교수님풀이 3-1 풀이 동일
```

```python
days_of_week(days_of_war)
```

```python
days_of_week(days_of_war).value_counts()

#교수님풀이 3-2
#days_of_war.dropna().apply(days_of_week).value_counts()
 
```
